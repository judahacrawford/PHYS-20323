```python
import numpy as np
class Particle(object):
    
    def __init__(self, mass, x,y,z,vx,vy,vz):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.vel = (np.sqrt(vx**2+vy**2+vz**2))
    
    def kine(self):
        print(0.5*self.mass*(self.vel)**2)
    
    def distance(self,other):
        print(np.sqrt((self.x-other.x)**2+(self.y-other.y)**2+(self.z-other.z)**2))
        
    def new_pos(self,other):
        dt = other
        print('x:',(self.x+self.vx*dt),',','y:',(self.y+self.vy*dt),',','z:',(self.z+self.vz*dt))
        return [(self.x+self.vx*dt),(self.y+self.vy*dt),(self.z+self.vz*dt)]
    
    
class ChargedParticle(Particle):
    
    def __init__(self, mass, x,y,z,vx,vy,vz,q):
        self.mass = mass
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        self.vel = (np.sqrt(vx**2+vy**2+vz**2))
        self.q = q
```

```python
p1 = Particle(1,0,0,0,1,2,7)
p2 = Particle(2,45,0,1,2,1,2)
p1.kine()
p1.distance(p2)
p1.new_pos(2)
p1 = ChargedParticle(1,0,0,0,1,2,1,6)
p1.kine()
p1.distance(p2)
```
